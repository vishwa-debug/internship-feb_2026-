```python
# Store 5 students and marks
students = {
    "A": 80,
    "B": 95,
    "C": 70,
    "D": 60,
    "E": 50
}

total = 0
topper = ""
highest = 0

for name in students:
    marks = students[name]
    total += marks

    # Find topper
    if marks > highest:
        highest = marks
        topper = name

    # Assign grade
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "D"

    print(name, ":", marks, "Grade:", grade)

# Calculate average
average = total / 5

print("Class Average:", average)
print("Topper:", topper)
```
